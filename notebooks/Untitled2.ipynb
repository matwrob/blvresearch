```python
import datetime as dt
import pandas as pd

from concat_overlord.dbpools import BLVPRICES_READ, FSDB
from blvprices.universe import Universe, Security
from blvutils.sql import sql_values
```

```python
def prepare_result_container(entities):
    perm_sec_ids = _get_perm_sec_ids(entities)
    securities = _get_security_objects(list(perm_sec_ids.values()))
    result = dict()
    for sec in securities:
        fs_index = _get_factset_index(sec.fs_perm_sec_id)
        correct_index = _add_correct_time(sec, fs_index)
        result[sec.factset_entity_id] = pd.TimeSeries(index=correct_index)
    return result


def _get_perm_sec_ids(entities):
    q = """SELECT factset_entity_id, fs_perm_sec_id
           FROM universe_item
           WHERE factset_entity_id = ANY({0})
        """.format(sql_values(entities))
    res = BLVPRICES_READ.execute(q, entities)
    result = dict()
    for t in res.fetchall():
        if t[0] not in result.keys():
            result[t[0]] = t[1]
        if t[1] != result[t[0]]:
            print('two different fs_perm_sec_ids?????')
    return result


def _get_security_objects(fs_ids):
    query = """SELECT * FROM universe_item
               WHERE universe_id = 'RESEARCH_2014'    
               AND fs_perm_sec_id = ANY({0})
            """.format(sql_values(fs_ids))
    res = BLVPRICES_READ.execute(query, fs_ids)
    return [Security(r) for r in res.fetchall()]


def _get_factset_index(fs_perm_sec_id):
    q = """SELECT date
           FROM fp_basic_bd
           WHERE fs_perm_sec_id = '%s'""" % fs_perm_sec_id
    res = FSDB.execute(q)
    return pd.DatetimeIndex([t[0] for t in res.fetchall()])


def _add_correct_time(security, date_index):
    time = security.exchange['close'].split(':')
    add_time = dt.timedelta(hours=int(time[0]), minutes=int(time[1]))
    new_index = date_index + add_time
    new_index = new_index.tz_localize(security.exchange['zone'])
    new_index.tz_convert('UTC')
    return new_index
```
