# Comparison of results for NEWS-based signals

```python
from concat_overlord.universes.standard import new_universe
from concat_overlord.dbpools import CONCAT_STANDARD_USER
from concat_overlord.update_client import make_output
from blvprices.universe import Universe
```

### Load CH-universe and data

```python
funi = Universe('CONCAT_STANDARD')
```

```python
output = make_output(Universe([s for s in funi if s.country == 'CH']))
```

### Import SignalReport

```python
from blvresearch.concat.signals.signal_testing import SignalReport
```

# Strategy 1

OVERVIEW:  
First identifies important events, i.e.  
(1) relevant news need to exist  
(2) alpha needs to exceed expected alpha by 1 standard deviation  
  
    Remark: alpha moments are calculated based on previous 6 months  
  
Next for each of these days:  
(3) checks what was the return on that day & invests accordingly:  
  
    i.e. if alpha was positive => buy,  
         if alpha was negative => sell
  

```python
from blvresearch.concat.signals.news_based.news_signals_strategy1 import get_signals
```

```python
signals = dict()
for k, v in output.items():
    if len(v) > 0:
        signals[k] = get_signals(v)
```

```python
sr = SignalReport(signals, output)
```

```python
print(sr.get())  # 22.07.2014
```
