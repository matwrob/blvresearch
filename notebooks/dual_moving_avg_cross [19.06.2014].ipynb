# Dual Moving Average Crossover

```python
from concat_overlord.universes.standard import new_universe
from concat_overlord.standard import make_output
from fsprices.universe import Universe
```

### Load CH-universe and data

```python
funi = new_universe()
```

```python
output = make_output(Universe([s for s in funi if s.country == 'CH']))
```

### Import SignalReport

```python
from blvresearch.concat.signals.signal_testing import SignalReport
```

### Import get_signals function

```python
from blvresearch.concat.signals.dual_moving_average_crossover import get_signals
```

### Pay attention to the difference between results based on cumulative alphas averaging and daily alphas averaging. When using daily alphas we get better results but number of signals increases, with the difference in a number of signals reaching the magnitude of 2 (twice more signals) for short windows. For longer averages it seems to change, i.e. better results are obtained when using cumulative alphas and not daily alphas.

# LW = 30 days, SW = 5 days, CW = 5 days

```python
long_window = 30
short_window = 5
confirmation_window = 5
```

```python
signals = dict()
for k, v in output.items():
    if len(v) > 0:
        signals[k] = get_signals(v, long_window, short_window, confirmation_window)
```

```python
sr = SignalReport(signals, output)
```

```python
print(sr.get())  # 19.06.2014  EMA calculated based on daily alphas
```

```python
print(sr.get())  # 19.06.2014  EMA calculated based on cumulative sum of alphas
```
