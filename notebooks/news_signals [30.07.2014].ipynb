# News + 0 days

OVERVIEW:  
First identifies important events, i.e days where:  
(1) relevant news was published  
(2) alpha on that day exceeds expected alpha by 2 standard deviations  

    Remarks:
    alpha moments are calculated based on previous 6 months of data  

Next for each of these days:  
(3) checks if alpha was positive/negative and invests accordingly  
  
    alpha > 0 => buy signal  
    alpha < 0 => sell signal  

```python
from blvresearch.concat.signals.news_based.news1day import get_news1day_signals
```

```python
signals = dict()
for k, v in output.items():
    if len(v) > 0:
        days_to_check, first_loc, last_loc = get_days_to_check(v)
        signals[k] = get_news1day_signals(days_to_check, v, first_loc, last_loc)
```

```python
sr = SignalReport(signals, output)
```

```python
print(sr.get())  # 04.08.2014  abnormal alpha is 2 sigmas away from the mean, with consecutive values
```

```python
print(sr.get())  # 04.08.2014  abnormal alpha is 2 sigmas away from the mean
```

```python
print(sr.get())  # 30.07.2014  abnormal alpha is 2 sigmas away from the mean
```

# News + CMO

OVERVIEW:  
First identifies important events, i.e days where:  
(1) relevant news was published  
(2) alpha on that day exceeds expected alpha by 2 standard deviations  
  
    Remarks:  
    alpha moments are calculated based on previous 6 months of data  
  
Next, for each of these days, checks:  
(3) alpha on that day  
(4) cumulative alpha over 3 days (that day + 2 next days)  
(5) if over the period starting 5 days before that day, and ending 2 days  
    after that day, a buy/sell signal - generated by CMO strategy -  
    occurred  
  
    alpha > 0 AND cum_alpha > 0 AND cmo_buy_signal => buy signal  
    alpha < 0 AND cum_alpha < 0 AND cmo_sell_signal => sell signal  

```python
from blvresearch.concat.signals.news_based.news_cmo import get_news_cmo_signals
```

```python
signals = dict()
for k, v in output.items():
    days_to_check, first_loc, last_loc = get_days_to_check(v)
    if len(v) > 0:
        signals[k] = get_news_cmo_signals(days_to_check, v, first_loc, last_loc)
```

```python
sr = SignalReport(signals, output)
```

```python
print(sr.get())  # 04.08.2014 without remove_consecutive_values function
```

```python
print(sr.get())  # 04.08.2014
```

```python
print(sr.get())  # 30.07.2014
```

# News + MA

OVERVIEW:  
First identifies important events, i.e.  
(1) relevant news need to exist  
(2) alpha needs to exceed expected alpha by 2 standard deviation  
  
    Remark: alpha moments are calculated based on previous 6 months  
  
Next for each of these days, it checks:  
(3) what was the return on that day  
(4) what was the return over 3 days (including signal day)  
(5) if over the previous 5 or the next 3 days a buy/sell signal - generated by MA strategy - occurred

```python
from blvresearch.concat.signals.news_based.news_movavg import get_news_movavg_signals
```

```python
signals = dict()
for k, v in output.items():
    days_to_check, first_loc, last_loc = get_days_to_check(v)
    if len(v) > 0:
        signals[k] = get_news_movavg_signals(days_to_check, v, first_loc, last_loc)
```

```python
sr = SignalReport(signals, output)
```

```python
print(sr.get())  # 04.08.2014 without remove_consecutive_values function
```

```python
print(sr.get())  # 04.08.2014
```

```python
print(sr.get())  # 30.07.2014
```
