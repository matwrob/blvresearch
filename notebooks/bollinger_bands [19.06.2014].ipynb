```python
import pandas as pd
```

```python
returns = data['alpha'].cumsum()
```

```python
mvg_std = pd.rolling_std(returns, window=20)
```

```python
exp_mvg_std = pd.ewmstd(returns, span=20)
```

```python
exp_mvg_std
```

```python
mvg_std.name = 'simple'
exp_mvg_std.name = 'exp'
```

```python
pd.concat([mvg_std, exp_mvg_std], axis=1).plot()
```

```python
bollinger_period = 20
```

```python
mvg_avg = pd.rolling_mean(returns, window=bollinger_period)
```

```python
lower_band = mvg_avg - 2 * mvg_std
upper_band = mvg_avg + 2 * mvg_std
```

```python
returns.name = 'return'
lower_band.name = 'lower_band'
upper_band.name = 'upper_band'
```

```python
pd.concat([returns, lower_band, upper_band], axis=1).plot()
```
