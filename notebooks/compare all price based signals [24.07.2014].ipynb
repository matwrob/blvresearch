# Comparison of results for PRICE-based signals

```python
from concat_overlord.universes.standard import new_universe
from concat_overlord.dbpools import CONCAT_STANDARD_USER
from concat_overlord.update_client import make_output
from blvprices.universe import Universe
```

### Load CH-universe and data

```python
funi = Universe('CONCAT_STANDARD')
```

```python
output = make_output(Universe([s for s in funi if s.country == 'CH']))
```

### Import SignalReport

```python
from blvresearch.concat.signals.signal_testing import SignalReport
```

# Moving Average

```python
from blvresearch.concat.signals.price_based.moving_average import get_movavg_signals
```

```python
signals = dict()
for k, v in output.items():
    if len(v) > 0:
        signals[k] = get_movavg_signals(v, window=20, confirmation_window=5)
```

```python
sr = SignalReport(signals, output)
```

```python
print(sr.get())  # 24.07.2014
```

```python
print(sr.get())  # 23.07.2014
```
