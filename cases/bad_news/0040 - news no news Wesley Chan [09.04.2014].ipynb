### Table 1: Summary of news observations in analysis, selected months, 2011-2013

```python
years = [2011, 2012, 2013]
```

##### Stocks traded

```python
def stock_was_traded(data_frame):
    days_with_returns = data_frame['abs_ret'].dropna()
    if len(days_with_returns) > 0:
        return True
    return False
```

```python
all_entities_traded = {year: list() for year in years}
```

```python
for year in years:
    start_date = str(year) + "-12-01"
    end_date = str(year) + "-12-31"    
    for entity_id, entity_data in DATA.items():
        if stock_was_traded(entity_data[start_date:end_date]):
            all_entities_traded[year].append(entity_id)
```

```python
for year in years:
    print(year, ':', len(all_entities_traded[year]))  # [09.04.2014]
```

##### Stocks with and without news

```python
def news_days_count(data_frame):
    days_with_news = data_frame['news'].dropna()
    return len(days_with_news)
```

```python
entities_with_news = {year: list() for year in years}
entities_without_news = {year: list() for year in years}
```

```python
for year in years:
    start_date = str(year) + "-12-01"
    end_date = str(year) + "-12-31"
    for entity_id in all_entities_traded[year]:
        d = DATA[entity_id][start_date:end_date]
        if news_days_count(d) > 0:
            entities_with_news[year].append(entity_id)
        else:
            entities_without_news[year].append(entity_id)
```

```python
for year in years:
    print(year, ':', len(entities_with_news[year]))  # [09.04.2014]
```

```python
for year in years:
    print(year, ':', len(entities_without_news[year]))  # [09.04.2014]
```

```python
for year in years:  # [09.04.2014]
    print(year, ':', len(all_entities_traded[year]))  # total count of entities traded
    print(len(entities_with_news[year]) + len(entities_without_news[year]))  # sum of entities with and without news
```

##### Stocks with news; by number of days with news in December (4 or fewer vs 5 or more)

```python
entities_with_4_or_fewer_news_days = {year: list() for year in years}
entities_with_5_or_more_news_days = {year: list() for year in years}
```

```python
for year in years:
    start_date = str(year) + "-12-01"
    end_date = str(year) + "-12-31"    
    for entity_id in entities_with_news[year]:
        d = DATA[entity_id][start_date:end_date]
        if news_days_count(d) > 4:
            entities_with_5_or_more_news_days[year].append(entity_id)
        else:
            entities_with_4_or_fewer_news_days[year].append(entity_id)
```

### 'News' portfolios

```python
cum_abs_ret_news_entities = {year: dict() for year in years}
```

```python
for year in years:
    start_date = str(year) + "-12-01"
    end_date = str(year) + "-12-31"
    for entity_id in entities_with_news[year]:
        car = DATA[entity_id]['abs_ret'][start_date:end_date].sum()
        cum_abs_ret_news_entities[year][entity_id] = car
```

```python
cum_abs_ret_news_entities = {
    k: pd.Series(v) for k, v in cum_abs_ret_news_entities.items()
}
```

```python
winners_news_entities = {y: list() for y in years}
losers_news_entities = {y: list() for y in years}
```

```python
return_breakpoints = {year: dict() for year in years}
```

```python
for year, series in cum_abs_ret_news_entities.items():
    portfolio_size = int(len(series) / 3)  # rounding to integer < value
    series.sort()
    winners_news_entities[year] = series[-portfolio_size:]
    losers_news_entities[year] = series[:portfolio_size]
    return_breakpoints[year]['winners'] = winners_news_entities[year][-portfolio_size:][0]
    return_breakpoints[year]['losers'] = losers_news_entities[year][:portfolio_size][-1]
```

```python
for year, series in winners_news_entities.items():  # [09.04.2014]
    print(year, ':', len(series))
```

```python
for year, series in losers_news_entities.items():  # [09.04.2014]
    print(year, ':', len(series))
```

```python
for year, breakpoints in return_breakpoints.items():  # [09.04.2014]
    print(year, ':', breakpoints['winners'])
```

```python
for year, breakpoints in return_breakpoints.items():  # [09.04.2014]
    print(year, ':', breakpoints['losers'])
```

### 'No-news' portfolios

```python
cum_abs_ret_no_news_entities = {year: dict() for year in years}
```

```python
for year in years:
    start_date = str(year) + "-12-01"
    end_date = str(year) + "-12-31"
    for entity_id in entities_without_news[year]:
        car = DATA[entity_id]['abs_ret'][start_date:end_date].sum()
        cum_abs_ret_no_news_entities[year][entity_id] = car
```

```python
cum_abs_ret_no_news_entities = {
    k: pd.Series(v) for k, v in cum_abs_ret_no_news_entities.items()
}
```

```python
winners_no_news_entities = {y: list() for y in years}
losers_no_news_entities = {y: list() for y in years}
```

```python
for year, series in cum_abs_ret_no_news_entities.items():
    series.sort()

    winners_bp = return_breakpoints[year]['winners']
    winners_no_news_entities[year] = series[series > winners_bp]

    losers_bp = return_breakpoints[year]['losers']
    losers_no_news_entities[year] = series[series < losers_bp]
```

```python
for year, series in winners_no_news_entities.items():  # [09.04.2014]
    print(year, ':', len(series))
```

```python
for year, series in losers_no_news_entities.items():  # [09.04.2014]
    print(year, ':', len(series))
```
