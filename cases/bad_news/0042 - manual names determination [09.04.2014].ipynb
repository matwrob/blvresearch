```python
import numpy as np
```

```python
from fsprices.universe import (
    Universe, Security, get_research_universe
)

from blvresearch.core.research_data import (
    get_research_data_by_country
)
```

```python
start_date = '2011-01-01'
end_date = '2013-12-31'
```

```python
def criteria(sec_meta):
    if (sec_meta['avg_vol'] > 1000 and
        sec_meta['market_cap'] > 1000000):
        return True
    return False
```

```python
DATA = get_research_data_by_country(
    ['US'], criteria, start_date, end_date
)
```

```python
UNI = get_research_universe(start_date[:4])
META = {v['factset_entity_id']: Security(v)
        for k, v in UNI.iterrows()}
```

```python
len(DATA)  # number of entities with data
```

##### Get 100 random news for each stock

```python
news_by_entity = dict()
for entity_id, entity_data in DATA.items():
    news_by_entity[entity_id] = list()
    news = entity_data['news'].dropna()
    for count in range(0, 100):
        index = np.random.random_integers(low=0, high=len(news) - 1)
        for n in news[index]:
            news_by_entity[entity_id].append(n)
            break
```

```python
len(news_by_entity)
```

```python
for entity_id, nl in news_by_entity.items():
    print(entity_id, len(nl))
```

```python
string = ""
for entity_id, nl in news_by_entity.items():
    string += entity_id + ('\n#######\n')
    for n in nl:
        string += n['raw_headline'] + '\n'
        string += n['clean_headline'] + '\n'
        string += n['body'].split('\n\n')[0]
        string += '\n\n'
    string += '------------------------------------\n\n\n\n'
```

```python
text_file = open('string_to_check_names_USA_1491ENT_2011_2013.txt', "w")
text_file.write(string)
text_file.close()
```
