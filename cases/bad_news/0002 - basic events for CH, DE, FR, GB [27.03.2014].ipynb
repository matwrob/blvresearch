```python
import numpy as np
```

```python
from fsprices.universe import Universe

from blvresearch.core.research_data import get_research_data_by_country
from blvresearch.core.event_study import EventDetector, EventList
from blvresearch.core.news_events import NewsEvent
from blvresearch.core.utils import mean_test
```

```python
start_date = '2011-01-01'
end_date = '2013-12-31'
```

```python
def criteria(sec_meta):
    if sec_meta['avg_vol'] > 1000 and sec_meta['market_cap'] > 1000000:
        return True
    return False
```

```python
DATA = get_research_data_by_country(['CH', 'DE', 'FR', 'GB'],
                                    criteria,
                                    start_date,
                                    end_date)
```

```python
UNI = Universe.research(int(start_date[:4]))
```

```python
len(DATA)  # number of entities with data
```

##### Event 1: many news & alpha < miu - 2 * sigma

```python
class ManyNewsAlphaSmallerThanMiuMinus2Sigmas(NewsEvent):
    def triggers(self):
        if self.has_many_news and self.alpha_is_smaller_than_miu_minus_2_sigmas:
            return True
        return False
```

```python
detector1 = EventDetector(ManyNewsAlphaSmallerThanMiuMinus2Sigmas)
```

```python
events1 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector1.run(entity_id, entity_data, UNI[entity_id])
    if len(entity_events) > 0:
        events1[entity_id] = entity_events
```

```python
list1 = EventList.from_dict(events1)
```

```python
def get_common_events(event_list1, event_list2):
    by_entity1 = event_list1.split_by_entity()
    by_entity2 = event_list2.split_by_entity()
    result = EventList()
    for entity_id, events1 in by_entity1.items():
        if entity_id in by_entity2.keys():
            dates2 = [e.date for e in by_entity2[entity_id]]
            for e in events1:
                if e.date in dates2:
                    result.append_event(e)
    return result
```

```python
cmn = get_common_events(list1, list6)
```

```python
mean_test(cmn, 'alpha', 20)
```

```python
def get_different_events(minuend_list, subtrahend_list):
    by_entity1 = minuend_list.split_by_entity()
    by_entity2 = subtrahend_list.split_by_entity()
    result = EventList()
    for entity_id, events1 in by_entity1.items():
        if entity_id in by_entity2.keys():
            dates2 = [e.date for e in by_entity2[entity_id]]
            for e in events1:
                if e.date not in dates2:
                    result.append_event(e)
        else:
            for e in events1:
                result.append_event(e)
    return result
```

```python
diff = get_different_events(list6, cmn)
```
