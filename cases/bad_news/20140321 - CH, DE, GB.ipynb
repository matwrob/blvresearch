```python
from fsprices.universe import Universe
from blvresearch.core.research_data import get_research_data
from blvresearch.core.news_events import NewsEvent
from blvresearch.core.event_study import EventDetector, EventList
from blvresearch.core.utils import mean_test
```

```python
start_date = '2011-01-01'
end_date = '2012-12-31'
year = int(start_date[:4])
```

```python
UNI = Universe.research(year)
```

```python
countries = ['CH', 'DE', 'FR', 'GB']
```

```python
entities = [s.factset_entity_id for s in UNI 
            if s.country in countries]
```

```python
DATA = get_research_data(entities, start_date, end_date)
```

```python
len(DATA)  # number of entities with price data & news
```

##### Event 1: day has many news and alpha < -2%

```python
class ManyNewsAlphaSmallerThanMinus2pct(NewsEvent):
    def triggers(self):
        if self.has_many_news and self.alpha_is_smaller_than_minus_2pct:
            return True
        return False
```

```python
detector1 = EventDetector(ManyNewsAlphaSmallerThanMinus2pct)
```

```python
events1 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector1.run(entity_id, entity_data, UNI[entity_id])
    if len(entity_events) > 0:
       events1[entity_id] = entity_events
```

##### Event 2: day has many news and alpha is smaller than miu minus 2 sigmas

```python
class ManyNewsAlphaSmallerThanMiuMinus2Sigmas(NewsEvent):
    def triggers(self):
        if self.has_many_news and self.alpha_is_smaller_than_miu_minus_2_sigmas:
            return True
        return False
```

```python
detector1 = EventDetector(ManyNewsAlphaSmallerThanMiuMinus2Sigmas)
```

```python
events1 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector1.run(entity_id, entity_data, UNI[entity_id])
    if len(entity_events) > 0:
        events1[entity_id] = entity_events
```

```python
len(events1)  # number of entities with at least 1 event
```

```python
event_list1 = EventList.from_dict(events1)
```

```python
event_list1
```

```python
mean_test(event_list1, 'alpha', 20)
```
