```python
from fsprices.universe import (
    Universe, Security, get_research_universe
)

from blvresearch.core.research_data import (
    get_research_data_by_country
)
from blvresearch.core.event_study import (
    EventDetector, EventList
)
from blvresearch.core.news_events import NewsEvent
from blvresearch.core.utils import (
    mean_test, plot_histogram
)
```

```python
start_date = '2011-01-01'
end_date = '2013-12-31'
```

```python
def criteria(sec_meta):
    if (sec_meta['avg_vol'] > 1000 and
        sec_meta['market_cap'] > 1000000):
        return True
    return False
```

```python
DATA = get_research_data_by_country(
    ['US'], criteria, start_date, end_date
)
```

```python
UNI = get_research_universe(start_date[:4])
META = {v['factset_entity_id']: Security(v)
        for k, v in UNI.iterrows()}
```

```python
len(DATA)  # number of entities with data [11.04.2014]
```

##### Event 1: many news & alpha < miu - 2 * sigma

```python
class ManyNewsAlphaSmallerThanMiuMinus2Sigmas(NewsEvent):
    def triggers(self):
        n_m = self.additional_data['news_count_miu']
        n_s = self.additional_data['news_count_sigma']
        a_m = self.additional_data['alpha_miu']
        a_s = self.additional_data['alpha_sigma']
        if (self.event_day_news_count > n_m + 2 * n_s and
            self.event_day_alpha < a_m - 2 * a_s):
            return True
        return False
```

```python
detector1 = EventDetector(ManyNewsAlphaSmallerThanMiuMinus2Sigmas)
```

```python
events1 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector1.run(entity_id, entity_data, META[entity_id])
    if len(entity_events) > 0:
        events1[entity_id] = entity_events
```

```python
len(events1)  # number of entities with at least 1 event [11.04.2014]
```

```python
list1 = EventList.from_dict(events1)
```

```python
list1  # total number of events found [11.04.2014]
```

```python
mean_test(list1, 'alpha', lag=0, length=1)  # test hypoth. mean return is zero [11.04.2014]
```

```python
def plot_histogram(event_list, attribute, lag=2, length=20):
    series = [e.concat_data.series_after(attribute, lag=2, length=length)
              for e in event_list]
    returns = [s.sum() for s in series]
    plt.hist(returns, bins=50, normed=True, alpha=0.6, color='g')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    mu, std = stats.norm.fit(returns)
    p = stats.norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)
    title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
    plt.title(title)
```
