# Basic events with news for USA

* years: 2011-2013
* universe: restricted to the biggest 1491 entities

Across 1491 entities and 3 full calendar years I'm looking for days characterized by:  
  
* abnormal return and abnormal news coverage

I check the average CARs (cumulative abnormal returns) before and after events I find.

### Events:

1. many news & alpha < miu - 2 * sigma
2. many news & alpha > miu + 2 * sigma
3. many news & alpha < miu - sigma
4. many news & alpha > miu + sigma
5. many news & miu - 2 * sigma < alpha < miu - sigma
6. many news & miu + sigma < alpha < miu + 2 * sigma

##### What does many news mean?

Many news means news count on event day is greater than mean news count + 2 std. deviations (mean & std calculated over 3 years).
This is a very crude way to determine outliers in news coverage. For example a company with equal amount of news every day and only
one day with slightly excess news count most likely won't match these criteria, and therefore we won't find any events.

### Event 1: many news & alpha < miu - 2 * sigma

The meaning of effects:  
    
* positive return after the event = reversal
* negative return after the event = drift

```python
class Event1(NewsEvent):
    @property
    def description(self):
        return 'nc > n_m + 2 * n_s & a < a_m - 2 * a_s'
    def triggers(self):
        n_m = self.additional_data['news_count_miu']
        n_s = self.additional_data['news_count_sigma']
        a_m = self.additional_data['alpha_miu']
        a_s = self.additional_data['alpha_sigma']
        if (self.event_day_news_count > n_m + 2 * n_s and
            self.event_day_alpha < a_m - 2 * a_s):
            return True
        return False
```

```python
detector1 = EventDetector(Event1)
```

```python
events1 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector1.run(entity_id, entity_data, META[entity_id])
    if len(entity_events) > 0:
        events1[entity_id] = entity_events
```

```python
len(events1)  # number of entities with at least 1 event [28.04.2014]
```

```python
list1 = EventList.from_dict(events1)
```

```python
list1  # total number of events found [28.04.2014]
```

##### One particular flow of this methodology

The above Apple, Inc. example shows one particular flow of this methodology, namely, I sometimes count returns multiple times. For example, I find two close events (6th & 7th of October, 2011), and therefore I will artificially increase the significance of any results that follow.
The events of 2011-10-06 and 2011-10-07 should be considered as one event and I should count it only once.

The question is how should I solve this problem.
The easiest way is to select only those events that are not followed by other events.  
  
This could be obtained by a 2-step selection procedure:  

First I should select all events that match the criteria (e.g. many news & alpha < miu - sigma).  
Second I should remove from this list all events that are:  

* followed (if I want to check effects afterwards) or  
* preceded (if I want to check effects before)  

by another events of this type.

```python
for d in events3['000C7F-E']:
    print(d)
    for n in d.concat_data.series_after('news', lag=0, length=1)[0]:
        print('\t\t', n['clean_headline'])
```

### Event 5: many news & miu - 2 * sigma < alpha < miu - sigma

Here I look at events where alpha on event day is below 1 sigma and above 2 sigmas to the left from the mean.  
  
    
We can think of such days as days, where something significant happened, but not as severe as in Event 1 study.
  
    
We can also think of this study as the analysis of the events that make up the difference between Event 3 and Event 1.

```python
class Event5(NewsEvent):
    @property
    def description(self):
        return 'n_c > n_m + 2 * n_s & a_m - 2 * a_s < a < a_m - a_s'
    def triggers(self):
        n_m = self.additional_data['news_count_miu']
        n_s = self.additional_data['news_count_sigma']
        a_m = self.additional_data['alpha_miu']
        a_s = self.additional_data['alpha_sigma']
        if (self.event_day_news_count > n_m + 2 * n_s and
            self.event_day_alpha > a_m - 2 * a_s and
            self.event_day_alpha < a_m - a_s):
            return True
        return False
```

```python
detector5 = EventDetector(Event5)
```

```python
events5 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector5.run(entity_id, entity_data, META[entity_id])
    if len(entity_events) > 0:
        events5[entity_id] = entity_events
```

##### Results

```python
res5 = tabular_results([list5], 'alpha', result_days)
```

```python
res5.T  # [28.04.2014]
```

The average return on event day (-2.39%) shows I'm analyzing less severe events here (compared to Event 1 study).  
  
These results can be compared to Event 1 results, because they don't have common events.  

I do not find significant negative return before Event 5, as it was the case with Event 1.  
For Event 1 we had -0.89% average CAR over 20 trading days before the event. Here I find -0.28% without statistical significance.  

On the other hand I find some significant returns after Event 5, whereas I haven't found after Event 1.  
For example, average CAR over 20 trading days after Event 5 is -0.68% (drift), whereas for Event 1 I had insignificant return of -0.148%.

##### What stands behind these differences?

Results of Event 1 and Event 5 show the difference in effects after "very severe" events and "not so severe" events (in terms of market reaction to news).  

What we know is the following:  
  
* "very severe" events are preceded by negative drifts (20 trading days before)
* "not so severe" events are not
* "very severe" events are not followed by any abnormal returns (up to 20 trading days after)
* "not so severe" events seem to experience slight drift over the next 10-20 trading days  
  
  
The fact, that "not so severe" events are followed by some drift would point to underreaction.  
Because we don't find this for "very severe" events, this can implicate that markets discount "very significant" surprises efficiently,  
but has problems in estimating the meaning of "less significant" events, that's why might see drift after Event 5.

##### This is not a significant result really.

A few lines below we can see, that significance of this result (-0.68% average CAR over 20 days after the event) disappears, after removing 10 observations (sample size decreases from 916 to 906 events).
