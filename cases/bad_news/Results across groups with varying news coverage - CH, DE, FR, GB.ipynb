```python
import pandas as pd
import numpy as np
import scipy.stats as stats

from concat.core.utils import load_object
```

```python
DATA = load_object('data_europe_1433_2011_2013.pickle')
UNI = load_object('uni_europe_1433_2011_2013.pickle')[2013]
```

```python
from blvresearch.core.news_events import NewsEvent
from blvresearch.core.event_study import EventDetector, EventList, CloseEvents
```

```python
from concat.data.blvindustry import EXCHANGES
# Switzerland, Germany, UK, France
reputable_exchanges = EXCHANGES['CH'] + EXCHANGES['DE'] + EXCHANGES['FR'] + EXCHANGES['GB']
```

```python
class ManyNewsAlphaSmallerThanMinus2pct(NewsEvent):
    def triggers(self):
        if self.has_many_news and self.alpha_is_smaller_than_minus_2pct:
            return True
        return False
```

```python
detect_many_news_alpha_smaller_than_minus_2pct = EventDetector(ManyNewsAlphaSmallerThanMinus2pct)
```

```python
events_many_news_alpha_smaller_than_minus_2pct = dict()
count = len(DATA)
for entity_id, entity_data in DATA.items():
    print(entity_id, count)
    count -= 1
    entity_meta = UNI[entity_id]
    if entity_meta.exchange in reputable_exchanges:
        events_many_news_alpha_smaller_than_minus_2pct[entity_id] = detect_many_news_alpha_smaller_than_minus_2pct.run(entity_id, entity_data, entity_meta)
```

```python
entities_with_events = {k: v for k, v in events_many_news_alpha_smaller_than_minus_2pct.items() if len(v) > 0}
```

```python
list_many_news_alpha_smaller_than_minus_2pct = EventList.from_dict(events_many_news_alpha_smaller_than_minus_2pct)
```

```python
list_many_news_alpha_smaller_than_minus_2pct
```

```python
from concat.core.news import NewsList
```

```python
total_news_by_entity = dict()
for entity_id, event_list in entities_with_events.items():
    news = DATA[entity_id]['news']
    news_count = news.apply(lambda x: len(x) if isinstance(x, NewsList) else 0)
    total_news_by_entity[entity_id] = news_count.sum()
```

```python
total_news_by_entity = pd.Series(total_news_by_entity)
```

```python
total_news_by_entity.sort()
```

```python
bin_size = len(total_news_by_entity) / 4
```

```python
bin_size
```

```python
news_coverage = dict()
news_coverage['smallest'] = total_news_by_entity[:147].index
news_coverage['medium'] = total_news_by_entity[147:294].index
news_coverage['high'] = total_news_by_entity[294:441].index
news_coverage['biggest'] = total_news_by_entity[441:].index
```

```python
events_by_news_coverage = dict()
events_by_news_coverage['smallest'] = {k: events_many_news_alpha_smaller_than_minus_2pct[k] for k in news_coverage['smallest']}
events_by_news_coverage['medium'] = {k: events_many_news_alpha_smaller_than_minus_2pct[k] for k in news_coverage['medium']}
events_by_news_coverage['high'] = {k: events_many_news_alpha_smaller_than_minus_2pct[k] for k in news_coverage['high']}
events_by_news_coverage['biggest'] = {k: events_many_news_alpha_smaller_than_minus_2pct[k] for k in news_coverage['biggest']}
```

```python
event_lists_by_news_coverage = {k: EventList.from_dict(v) for k, v in events_by_news_coverage.items()}
```
