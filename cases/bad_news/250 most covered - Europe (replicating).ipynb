##### Event type: Many news and alpha < -2% but distance is not considered; all events are added no matter how close they are to each other

```python
class ManyNewsAlphaSmallerThanMinus2pct(NewsEvent):
    def triggers(self):
        if self._has_many_news() and self._alpha_smaller_than_minus_2pct():
            return True
        return False
```

```python
detect_mnastm2 = EventDetector(ManyNewsAlphaSmallerThanMinus2pct)
```

```python
events_mnastm2 = dict()
count = 250
for entity_id in EUROPE_HIGH_NEWS_COVERAGE_250:
    print(entity_id, count)
    count -= 1
    entity_data = DATA[entity_id]
    entity_meta = UNI[entity_id]
    events_mnastm2[entity_id] = detect_mnastm2.run(entity_id, entity_data, entity_meta)
```

```python
list_mnastm2 = EventList.from_dict(events_mnastm2)
```

```python
list_mnastm2
```

```python
mean_test(list_mnastm2, 'alpha', 20)
```

##### Event type: many news and alpha < -2% with LOOK-AHEAD: the next event with absolute alpha > 2% cannot occur within the following 30 days

```python
class ManyNewsAbsoluteAlphaGreaterThan2pct(NewsEvent):
    def triggers(self):
        if self._has_many_news() and self._absolute_alpha_greater_than_2pct():
            return True
        return False


class ManyNewsAlphaSmallerThanMinus2pctWithLookahead(NewsEvent):
    def triggers(self):
        if self._has_many_news() and self._alpha_smaller_than_minus_2pct():
            ce = CloseEvents(self)
            events = ce.find_after(
                ManyNewsAbsoluteAlphaGreaterThan2pct,
                self.DISTANCE_THRESH
            )
            if not events:
                return True
        return False
```

```python
detect_mnastm2wLA = EventDetector(ManyNewsAlphaSmallerThanMinus2pctWithLookahead)
```

```python
events_mnastm2wLA = dict()
count = 250
for entity_id in EUROPE_HIGH_NEWS_COVERAGE_250:
    print(entity_id, count)
    count -= 1
    entity_data = DATA[entity_id]
    entity_meta = UNI[entity_id]
    events_mnastm2wLA[entity_id] = detect_mnastm2wLA.run(entity_id, entity_data, entity_meta)
```

```python
list_mnastm2wLA = EventList.from_dict(events_mnastm2wLA)
```

```python
list_mnastm2wLA
```

```python
mean_test(list_mnastm2wLA, 'alpha', 20)
```
