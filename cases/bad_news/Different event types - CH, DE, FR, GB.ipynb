##### Event type: few news & alpha < miu - 2 sigmas, without considering distance, we take all events

```python
class FewNewsAlphaSmallerThanMiuMinus2Sigmas(NewsEvent):
    def triggers(self):
        if self.has_few_news and self.alpha_is_smaller_than_miu_minus_2_sigmas:
            return True
        return False
```
