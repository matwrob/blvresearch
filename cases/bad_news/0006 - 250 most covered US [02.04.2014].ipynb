```python
import pandas as pd
import numpy as np
```

```python
from fsprices.universe import Universe, Security, get_research_universe
from memnews.core import NewsList

from blvresearch.core.research_data import get_research_data_by_country
from blvresearch.core.event_study import EventDetector, EventList
from blvresearch.core.news_events import NewsEvent
from blvresearch.core.utils import mean_test
```

```python
start_date = '2011-01-01'
end_date = '2013-12-31'
```

```python
def criteria(sec_meta):
    if sec_meta['avg_vol'] > 1000 and sec_meta['market_cap'] > 1000000:
        return True
    return False
```

```python
DATA = get_research_data_by_country(['US'],
                                    criteria,
                                    start_date,
                                    end_date)
```

```python
UNI = get_research_universe(start_date[:4])
META = {v['factset_entity_id']: Security(v) for k, v in UNI.iterrows()}
```

```python
len(DATA)  # number of entities with data
```

##### Finding most news-covered entities

```python
def news_count_func(x):
    return len(x) if isinstance(x, NewsList) else 0
```

```python
news_by_entity = dict()
for entity_id, entity_data in DATA.items():
    news = entity_data['news'].map(news_count_func)
    news_by_entity[entity_id] = news.sum()
```

```python
news_by_entity = pd.Series(news_by_entity)
news_by_entity.sort()
```

```python
most_covered_250 = list(news_by_entity[-250:].index)
```

```python
most_covered_250[-1] == '000C7F-E'
```

##### Event 1: many news & alpha < miu - 2 * sigma

```python
class ManyNewsAlphaSmallerThanMiuMinus2Sigmas(NewsEvent):
    def triggers(self):
        if self.has_many_news and self.alpha_is_smaller_than_miu_minus_2_sigmas:
            return True
        return False
```

```python
detector1 = EventDetector(ManyNewsAlphaSmallerThanMiuMinus2Sigmas)
```

```python
events1 = dict()
for entity_id in most_covered_250:
    entity_events = detector1.run(entity_id, DATA[entity_id], META[entity_id])
    if len(entity_events) > 0:
        events1[entity_id] = entity_events
```

```python
len(events1)  # number of entities with at least 1 event
```

```python
list1 = EventList.from_dict(events1)
```

```python
list1
```

```python
mean_test(list1, 'alpha', 20)
```

```python
np.mean([e.concat_data.series_after('alpha', lag=0, length=1) for e in list1])
```

```python
import scipy.stats as stats
```

```python
def print_summary_wo_outliers(event_list, outlier_abs_threshold):
    attribute, length = 'alpha', 20
    series = [e.concat_data.series_after(attribute, lag=2, length=length)
              for e in event_list]
    returns = [s.sum() for s in series]
    print('Full sample')
    print('Size:', len(returns))
    print('Mean:', np.mean(returns))
    print(stats.ttest_1samp(returns, 0)[1])
    print()
    wo_outliers = [r for r in returns if abs(r) < outlier_abs_threshold]
    print('Without outliers')
    print('Size: ', len(wo_outliers))
    print('Mean:', np.mean(wo_outliers))
    print(stats.ttest_1samp(wo_outliers, 0)[1])
    

def plot_histogram_wo_outliers(event_list, outlier_abs_threshold):
    attribute, length = 'alpha', 20
    series = [e.concat_data.series_after(attribute, lag=2, length=length)
              for e in event_list]
    returns = [s.sum() for s in series
               if abs(s.sum()) < outlier_abs_threshold]
    plt.hist(returns, bins=50, normed=True, alpha=0.6, color='g')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    mu, std = stats.norm.fit(returns)
    p = stats.norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)
    title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
    plt.title(title)
```
