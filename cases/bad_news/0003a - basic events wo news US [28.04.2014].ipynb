# Basic events without news for USA

* years: 2011-2013
* universe: restricted to the biggest 1491 entities

Across 1491 entities and 3 full calendar years I'm looking for days characterized by:  
  
* abnormal return and no news

I check the average CARs (cumulative abnormal returns) before and after events I find.

### Events:

1. no news & alpha < miu - 2 * sigma
2. no news & alpha > miu + 2 * sigma
3. no news & alpha < miu - sigma
4. no news & alpha > miu + sigma
5. no news & miu - 2 * sigma < alpha < miu - sigma
6. no news & miu + sigma < alpha < miu + 2 * sigma

##### What does no news mean?

No news means news count on event day is zero

### Event 1: no news & alpha < miu - 2 * sigma

```python
class Event1(NewsEvent):
    @property
    def description(self):
        return 'n_c = 0 & a < a_m - 2 * a_s'
    def triggers(self):
        a_m = self.additional_data['alpha_miu']
        a_s = self.additional_data['alpha_sigma']
        if (self.event_day_news_count == 0 and
            self.event_day_alpha < a_m - 2 * a_s):
            return True
        return False
```

```python
detector1 = EventDetector(Event1)
```

```python
events1 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector1.run(entity_id, entity_data, META[entity_id])
    if len(entity_events) > 0:
        events1[entity_id] = entity_events
```

```python
len(events1)  # number of entities with at least 1 event [28.04.2014]
```

```python
list1 = EventList.from_dict(events1)
```

```python
list1  # total number of events found [28.04.2014]
```
