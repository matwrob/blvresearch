```python
from blvresearch.core.event_study import EventDetector
from blvresearch.core.news_events import NewsEvent
from fsprices.universe import Universe
from blvresearch.core.research_data import get_research_data_by_country
```

```python
eu_countries = ['CH', 'DE', 'FR', 'GB']
```

```python
UNI = Universe.research(2011)
DATA = get_research_data_by_country(eu_countries, '2011-01-01', '2013-12-31')
```

```python
class ManyNewsAlphaNegative(NewsEvent):
    def triggers(self):
        if self.has_many_news and self.alpha_is_negative:
            return True
        return False
```

```python
entity_id = '05HXPQ-E'
entity_data = DATA[entity_id]
```

```python
det = EventDetector(ManyNewsAlphaNegative)
```

```python
initial_list = det._create_initial_list(entity_id, entity_data, UNI[entity_id])
```
