```python
from blvworker.news.important_days import _get_mean_and_sigma


class NewsEvent(NewsEvent):
    def _get_news_length_moments(self):
        year = self.date.year
        start, end = pd.datetime(year, 1, 1), pd.datetime(year, 12, 31)
        news = self._data['news'][start:end]
        news_len = news.map(self._get_news_list_len)
        return _get_mean_and_sigma(news_len)
```

```python
def plot_histogram(values):
    plt.hist(values, bins=10, normed=True, alpha=0.6, color='g')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    mu, std = stats.norm.fit(values)
    p = stats.norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)
    title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
    plt.title(title)
```
