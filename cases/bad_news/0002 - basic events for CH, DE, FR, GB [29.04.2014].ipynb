### Event 1: ncount > ncount_miu + 2 * ncount_sigma AND alpha < alpha_miu - 2 * alpha_sigma

The meaning of effects:  
    
* positive return after the event = reversal
* negative return after the event = drift

```python
class Event1(NewsEvent):
    @property
    def description(self):
        return 'nc > n_m + 2 * n_s AND a < a_m - 2 * a_s'
    def triggers(self):
        n_m = self.additional_data['news_count_miu']
        n_s = self.additional_data['news_count_sigma']
        a_m = self.additional_data['alpha_miu']
        a_s = self.additional_data['alpha_sigma']
        if (self.event_day_news_count > n_m + 2 * n_s and
            self.event_day_alpha < a_m - 2 * a_s):
            return True
        return False
```

```python
detector1 = EventDetector(Event1)
```

```python
events1 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector1.run(entity_id, entity_data, META[entity_id])
    if len(entity_events) > 0:
        events1[entity_id] = entity_events
```

```python
len(events1)  # number of entities with at least 1 event [28.04.2014]
```

```python
list1 = EventList.from_dict(events1)
```

```python
list1  # [28.04.2014]
```

##### Event 1: UBS AG

```python
for d in events1['002ZDR-E']:
    print(d)
```

For UBS AG we have:
    
* 2011-08-02 - Switzerland & Germany will sign tax deal
* 2011-09-15 - Adoboli
* 2012-07-31 - Q2 results
* 2013-10-29 - Q3 results

```python
for d in events1['002ZDR-E']:
    print(d)
    for n in d.concat_data.series_after('news', lag=0, length=1)[0]:
        print('\t\t', n['clean_headline'])
```

##### Results

```python
res1 = tabular_results([list1], 'alpha', result_days)
```

```python
res1.T  # [28.04.2014]
```

As in USA case, we find statistically significant negative CAR over 20 trading days before the event.
No significant abnormal returns occur after events.
