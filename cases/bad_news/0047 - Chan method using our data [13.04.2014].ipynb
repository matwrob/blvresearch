```python
import pandas as pd
```

```python
from fsprices.universe import (
    Universe, Security, get_research_universe
)

from blvresearch.core.research_data import (
    get_research_data_by_country
)
from blvresearch.core.chan import (
    ChanNewsReturns, ChanNoNewsReturns
)
from blvresearch.core.std_rolling_portfolio import (
    JegadeeshReturns
)
```

```python
start_date = '2011-01-01'
end_date = '2013-12-31'
```

```python
def criteria(sec_meta):
    if (sec_meta['avg_vol'] > 1000 and
        sec_meta['market_cap'] > 1000000):
        return True
    return False
```

```python
DATA = get_research_data_by_country(
    ['CH', 'DE', 'FR', 'GB'], criteria, start_date, end_date
)
```

```python
len(DATA)  # total number of entities [13.04.2014]
```

```python
len(DATA)  # total number of entities [13.04.2014]
```

##### Check if returns with news are naned by ChanNoNewsReturns

##### Classic Jegadeesh Titman 1M-1M

```python
jt = JegadeeshReturns(DATA,
                      type_of_return='alpha',
                      no_of_quantiles=3)
```

```python
w, l, wml = (jt.get_winners_returns('M', test_per=1, hold_per=1),
             jt.get_losers_returns('M', test_per=1, hold_per=1),
             jt.get_wml_returns('M', test_per=1, hold_per=1))
w.name = 'W'
l.name = 'L'
wml.name = 'WmL'
```

```python
w_l_wml = pd.concat([w, l, wml], axis=1)
```

```python
w_l_wml.cumsum().plot()
```
