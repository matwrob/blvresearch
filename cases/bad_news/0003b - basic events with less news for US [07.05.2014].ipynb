# Basic events for USA

## ncount > ncount_miu + ncount_sigma

years: 2011-2013  
universe: restricted to the biggest 1491 entities

### Events:

1. ncount > ncount_miu + ncount_sigma AND alpha < alpha_miu - 2 * alpha_sigma
2. ncount > ncount_miu + ncount_sigma AND alpha > alpha_miu + 2 * alpha_sigma
3. ncount > ncount_miu + ncount_sigma AND alpha < alpha_miu - alpha_sigma
4. ncount > ncount_miu + ncount_sigma AND alpha > alpha_miu + alpha_sigma

### Event 1b: ncount > ncount_miu + ncount_sigma AND alpha < alpha_miu - 2 * alpha_sigma

```python
class Event1b(NewsEvent):
    @property
    def description(self):
        return 'nc > nc_m + nc_s AND a < a_m - 2 * a_s'
    def triggers(self):
        n_m = self.additional_data['news_count_miu']
        n_s = self.additional_data['news_count_sigma']
        a_m = self.additional_data['alpha_miu']
        a_s = self.additional_data['alpha_sigma']
        if (self.event_day_news_count > n_m + n_s and
            self.event_day_alpha < a_m - 2 * a_s):
            return True
        return False
```

```python
detector1b = EventDetector(Event1b)
```

```python
events1b = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector1b.run(entity_id, entity_data, META[entity_id])
    if len(entity_events) > 0:
        events1b[entity_id] = entity_events
```
