##### Event type: many news & alpha < -2%, without considering distance, we take all events

```python
class ManyNewsAlphaSmallerThanMinus2pct(NewsEvent):
    def triggers(self):
        if self.has_many_news and self.alpha_is_smaller_than_minus_2pct:
            return True
        return False
```

```python
detect_many_news_alpha_smaller_than_minus_2pct = EventDetector(ManyNewsAlphaSmallerThanMinus2pct)
```

```python
events_many_news_alpha_smaller_than_minus_2pct = dict()
count = len(DATA)
for entity_id, entity_data in DATA.items():
    print(entity_id, count)
    count -= 1
    entity_meta = UNI[entity_id]
    if entity_meta.exchange in reputable_exchanges:
        events_many_news_alpha_smaller_than_minus_2pct[entity_id] = detect_many_news_alpha_smaller_than_minus_2pct.run(entity_id, entity_data, entity_meta)
```

```python
len({k: v for k, v in events_many_news_alpha_smaller_than_minus_2pct.items() if len(v) > 0})
```

```python
list_many_news_alpha_smaller_than_minus_2pct = EventList.from_dict(events_many_news_alpha_smaller_than_minus_2pct)
```

```python
list_many_news_alpha_smaller_than_minus_2pct
```

```python
by_country = dict()
by_sector = dict()
for event in list_many_news_alpha_smaller_than_minus_2pct:
    country, sector = event.meta_data.country, event.meta_data.sector
    tmp_country = by_country.get(country, EventList())
    tmp_country.append_event(event)
    by_country[country] = tmp_country
    tmp_sector = by_sector.get(sector, EventList())
    tmp_sector.append_event(event)
    by_sector[sector] = tmp_sector
```
