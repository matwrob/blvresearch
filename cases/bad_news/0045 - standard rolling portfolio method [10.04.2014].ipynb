```python
def get_dataframe_of_returns(research_data, type_of_return='abs_ret',
                             freq=None):
    result = dict()
    for entity_id, df in research_data.items():
        result[entity_id] = df[type_of_return]
    result = pd.DataFrame.from_dict(result)
    if freq:
        return result.resample(freq, how=sum)
    return result
```

```python
def get_winners(df_of_returns, no_of_quantiles=3):
    s = _get_series_of_sorted_lists(df_of_returns)
    result = dict()
    for date, _list in s.items():
        size = int(len(_list) / no_of_quantiles)
        result[date] = _list[:size]
    return pd.Series(result)


def get_losers(df_of_returns, no_of_quantiles=3):
    s = _get_series_of_sorted_lists(df_of_returns)
    result = dict()
    for date, _list in s.items():
        size = int(len(_list) / no_of_quantiles)
        result[date] = _list[-size:]
    return pd.Series(result)


def _get_series_of_sorted_lists(df):
    result = dict()
    for date, series in df.iterrows():
        series = series.dropna()
        series.sort(ascending=False)
        result[date] = list(series.index)
    return pd.Series(result)
```

```python
def _get_series_of_sorted_lists(df, test_periods):
    result = dict()
    for i in range(len(df[:-test_periods])):
        frame = df[i:i + test_periods].dropna()
        series = frame.sum()
        series.sort(ascending=False)
        date = df.index[i + test_periods - 1]
        result[date] = list(series.index)
    return pd.Series(result)
```

```python
def get_ranked_securities(df, test_periods=1):
    result = dict()
    for i in range(len(df[:-test_periods])):
        frame = df[i:i + test_periods].dropna()
        series = frame.sum()
        series.sort(ascending=False)
        date = df.index[i + test_periods - 1]
        result[date] = series
    return result
```
