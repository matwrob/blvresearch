```python
runi2011 = Universe.research(2011)
```

```python
all_entities = [s.factset_entity_id for s in runi2011]
```

```python
eu_countries = ['AT', 'BE', 'CH', 'DE', 'DK', 'ES', 'FI', 'FR', 'GB', 'GR', 'IE', 'IT', 'NL', 'NO', 'PL', 'RU', 'SE', 'TR', 'ZA']
```

```python
eu_entities = [s.factset_entity_id for s in runi2011 if s.country in eu_countries]
```

```python
us_entities = [s.factset_entity_id for s in runi2011 if s.country == 'US']
```

```python
us_eu_entities = us_entities + eu_entities
```
