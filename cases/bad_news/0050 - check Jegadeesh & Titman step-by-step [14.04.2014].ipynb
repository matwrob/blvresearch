```python
sorted_returns = dict()
for i in range(len(returns) - test_per + 1):
    frame = returns[i:i + test_per]
    series = frame.sum(skipna=False).dropna()
    series.sort()  # ascending sort, losers on top
    date = returns.index[i + test_per - 1]
    sorted_returns[date] = series
```
