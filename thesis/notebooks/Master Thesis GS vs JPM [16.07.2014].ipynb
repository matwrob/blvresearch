```python
def get_absolute_cooccurrence(entity_id, data):
    other_entities = list(set(data.keys()) - set([entity_id]))
    entity_data = data[entity_id]
    result = pd.DataFrame(data=0,
                          index=entity_data.index,
                          columns=other_entities)
    for date, news_list in entity_data['news'].dropna().items():
        for news in news_list:
            common = list(set(news['entities']) &
                          set(other_entities))
            if common:
                result.loc[date][common] += 1
    return result
```

```python
coocc_goldman = get_absolute_cooccurrence(entity1, DATA)
```

```python
def get_return_correlations_for_2_entities(entity1, entity2, data, window=30):
    ret1, ret2 = data[entity1]['abs_ret'], data[entity2]['abs_ret']
    return pd.rolling_corr(ret1, ret2, window=window)
```

```python
retcorr_gs_jpm = get_return_correlations_for_2_entities(entity1, entity2, DATA)
```
