```python
def get_dickey_fuller_test_results(series):
    res = ts.adfuller(series)
    s = "Test statistic: %s\n" % res[0]
    s += "MacKinnon's approximate p-value: %s\n" % res[1]
    s += "No. of lags used: %s\n" % res[2]
    s += "No. of obs. used for ADF regression: %s\n" % res[3]
    s += "Critical values:\n"
    s += "* 1pct: %s\n" % res[4]['1%']
    s += "* 5pct: %s\n" % res[4]['5%']
    s += "* 10pct: %s" % res[4]['10%']
    print(s)
    return res
```

```python
res = get_dickey_fuller_test_results(corr)
```
