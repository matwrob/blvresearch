# Load return data

```python
PATH = '/vagrant/blvresearch/thesis/raw_data/djia_2010_2013/returns_wo_news'
returns = dict()
for file_name in os.listdir(PATH):
    name = file_name[:6] + "-E"
    returns[name] = pd.DataFrame.from_csv(PATH + "/" + file_name)
```

# Load news data

```python
import json
```

```python
PATH = '/vagrant/blvresearch/thesis/raw_data/djia_2010_2013/news'
news = dict()
for file_name in os.listdir(PATH):
    name = file_name[5:11] + "-E"
    tmp = pd.Series.from_csv(PATH + "/" + file_name, sep=";")
    news[name] = tmp.apply(lambda x: json.loads(x))
```

```python
news['000BY7-E']
```

# Calculate absolute cooccurrence

```python
abs_cooccurrence = dict()
```

```python
for entity_id, series in news.items():
    other_entities = list(set(news.keys()) - set([entity_id]))
    result = pd.DataFrame(data=0,
                          index=news[entity_id].index,
                          columns=other_entities)
    for date, news_dict in news[entity_id].iteritems():
        for k, v in news_dict.items():
            common = list(set(v['entities']) &
                          set(other_entities))
            if len(common) > 0:
                result.loc[date][common] += 1
    abs_cooccurrence[entity_id] = result
```

```python
abs_cooccurrence['000BY7-E']
```

# Calculate return correlations

```python
return_correlations = dict()
```

```python
WINDOW = 30
```

```python
for entity_id, df in returns.items():
    ret1 = df['abs_ret']
    other_entities = list(set(returns.keys()) - set([entity_id]))
    result = pd.DataFrame(columns=other_entities)
    for entity2 in other_entities:
        ret2 = returns[entity2]['abs_ret']
        result[entity2] = pd.rolling_corr(ret1, ret2, window=WINDOW)
    return_correlations[entity_id] = result
```

```python
return_correlations['000BY7-E'].dropna()
```

# Calculate return "magnitude" correlations

```python
magnitude_correlations = dict()
```

```python
for entity_id, df in returns.items():
    ret1 = df['abs_ret']
    other_entities = list(set(returns.keys()) - set([entity_id]))
    result = pd.DataFrame(columns=other_entities)
    for entity2 in other_entities:
        ret2 = returns[entity2]['abs_ret']
        result[entity2] = pd.rolling_corr(abs(ret1), abs(ret2), window=WINDOW)
    magnitude_correlations[entity_id] = result
```

```python
magnitude_correlations['000BY7-E'].dropna()
```

# Augmented Dickey-Fuller Test for non-stationarity of cooccurrence time series

```python
import statsmodels.tsa.stattools as ts
```

```python
for main_entity, df in abs_cooccurrence.items():
    print('Entity: ', main_entity, '\n')
    for entity_id, series in df.iteritems():
        result = ts.adfuller(series)

        stat = result[0]
        critical = result[4]

        if stat > critical['1%']:
            print(entity_id)
            print('non_stationary 1%')

        if stat > critical['5%']:
            print(entity_id)
            print('non_stationary 5%')

        if stat > critical['10%']:
            print(entity_id)
            print('non_stationary 10%')
```

```python
ts.adfuller(abs_cooccurrence['000BY7-E']['000Y55-E'])
```

================================  
For the series of absolute cooccurrence of entities:  
* '000BY7-E' (American Express Co.) and '000Y55-E' (United Technologies Corp.)  
  
at the 99/95/90% level the null hypothesis of non-stationarity cannot be rejected

  
All other series are stationary  

# Augmented Dickey-Fuller Test for non-stationarity of return correlation time series

```python
for main_entity, df in return_correlations.items():
    print('Entity: ', main_entity, '\n')
    for entity_id, series in df.iteritems():
        result = ts.adfuller(series.dropna())

        stat = result[0]
        critical = result[4]

        if stat > critical['1%']:
            print(entity_id)
            print('non_stationary 1%')

        if stat > critical['5%']:
            print(entity_id)
            print('non_stationary 5%')

        if stat > critical['10%']:
            print(entity_id)
            print('non_stationary 10%')
```

==========================================================  
For time series of return correlation between entities:  
    * '002615-E' (The Goldman Sachs Group, Inc.) & '05MDRV-E' (Visa, Inc.)  
    * '002Q0G-E' (Exxon Mobil Corp.) & '000DP6-E' (Chevron Corp.)  
    * '000DP6-E' (Chevron Corp.) & '000RBW-E' (NIKE, Inc.)  
    * '000CYC-E' (The Boeing Co.) & '000Q07-E' (Microsoft Corp.)  
  
at the 99% level the null hypothesis of non-stationarity cannot be rejected
