```python
import pandas as pd
import numpy as np
import scipy.stats as stats

from concat.core.utils import load_object
```

```python
UNI = load_object('uni_usa_2453_2011_2013.pickle')
UNI = UNI[2013]
DATA = load_object('data_usa_2453_2011_2013.pickle')
```

```python
from blvresearch.projects.event_study.news_events import NewsEvent
from blvresearch.core.es import EventDetector, EventList
```

```python
class ManyNewsAbsoluteAlphaGreaterThan2pct(NewsEvent):
    def triggers(self):
        if self._has_many_news() and self._absolute_alpha_greater_than_2pct():
            return True
        return False
```

```python
detect_mnaagt2 = EventDetector(ManyNewsAbsoluteAlphaGreaterThan2pct)
```

```python
events_mnaagt2 = dict()
for entity_id, entity_data in DATA.items():
    entity_meta = UNI[entity_id]
    events_mnaagt2[entity_id] = detect_mnaagt2.run(entity_id, entity_data, entity_meta)
```

```python
list_mnaagt2 = EventList.from_dict(events_mnaagt2)
```

```python
list_mnaagt2
```

```python
list_mnastm2 = EventList()
[list_mnastm2.append_event(e) for e in list_mnaagt2 if e.concat_data.series_after('alpha', lag=0, length=1) < 0]
```

```python
list_mnastm2
```

```python
negative_by_entity = list_mnastm2.split_by_entity()
all_by_entity = list_mnaagt2.split_by_entity()
```

```python
wo_followers = dict()
for entity_id, event_list in all_by_entity.items():
    if entity_id in negative_by_entity.keys():
        negative = negative_by_entity[entity_id]
        dates = [e.date for e in event_list]
        deltas = [(dates[i], (d - dates[i]).days) for i, d in enumerate(dates[1:])]
        filtered_dates = [t[0] for t in deltas if t[1] > 30]
        if event_list[len(event_list) - 1].concat_data.series_after('alpha', lag=0, length=1)[0] < 0:
            filtered_dates.append(dates[-1])
        tmp = EventList()
        [tmp.append_event(e) for e in negative if e.date in filtered_dates]
        wo_followers[entity_id] = tmp
```

```python
filtered_mnastm2_wo_followers = EventList.from_dict(wo_followers)
```
