```python
start_date = '2011-01-01'
end_date = '2013-12-31'
```

```python
def criteria(sec_meta):
    if (sec_meta['avg_vol'] > 1000 and
        sec_meta['market_cap'] > 1000000):
        return True
    return False
```

```python
DATA = get_research_data_by_country(
    ['US'], criteria, start_date, end_date
)
```

```python
len(DATA)  # total number of entities [15.04.2014]
```

### Jegadeesh & Titman 1M-1M (terciles)

I sort all stocks by their return in a given month, I select top tercile as winners and bottom tercile as losers (W. Chan uses terciles too). I check performance of 3 portfolios:  
  
1. winners  
2. losers  
3. short-long portfolio winners-minus-losers

```python
jt = JegadeeshTitman(DATA,
                     type_of_return='alpha',
                     no_of_quantiles=3)
```

```python
w = jt.get_winners_returns('M', test_per=1, hold_per=1)
w.name = 'W'

l = jt.get_losers_returns('M', test_per=1, hold_per=1)
l.name = 'L'

wml = jt.get_wml_returns('M', test_per=1, hold_per=1)
wml.name = 'WmL'
```

```python
w_l_wml = pd.concat([w, l, wml], axis=1)
```

```python
w_l_wml.cumsum().plot()
```
