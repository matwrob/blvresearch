```python
import pandas as pd
import numpy as np
import scipy.stats as stats

from concat.core.utils import load_object
```

```python
DATA = load_object('/vagrant/blvresearch/cases/data_europe_1433_2011_2013.pickle')
UNI = load_object('/vagrant/blvresearch/cases/uni_europe_1433_2011_2013.pickle')[2013]
```

```python
from blvresearch.core.news_events import NewsEvent
from blvresearch.core.event_study import EventDetector, EventList
from blvresearch.core.utils import mean_test
```

```python
from fsprices.universe import EXCHANGES
reputable_exchanges = [k for k, v in EXCHANGES.items()
                       if v['country'] in ['CH', 'DE', 'FR', 'GB']]
```

```python
class ManyNewsAlphaSmallerThanMiuMinus2Sigmas(NewsEvent):
    def triggers(self):
        if self.has_many_news and self.alpha_is_smaller_than_miu_minus_2_sigmas:
            return True
        return False
```

```python
detector1 = EventDetector(ManyNewsAlphaSmallerThanMiuMinus2Sigmas)
```

```python
events1 = dict()
count = len(DATA)
for entity_id, entity_data in DATA.items():
    print(entity_id, count)
    count -= 1
    if UNI[entity_id].exchange in reputable_exchanges:
        entity_events = detector1.run(entity_id, entity_data, UNI[entity_id])
        if len(entity_events) > 0:
            events1[entity_id] = entity_events
```
