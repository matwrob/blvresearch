##### Loading basic objects for analysis of news events

```python
from blvresearch.projects.event_study.news_events import NewsEvent, mean_test, plot_histogram_of_alphas
from blvresearch.core.es import EventDetector, EventList, create_dataframe_of_daily_effects
from blvresearch.projects.event_study.news_events import (
    ManyNewsAlphaGreaterThan2pct, ManyNewsAlphaSmallerThanMinus2pct, NoNewsAlphaGreaterThan2pct,
    NoNewsAlphaSmallerThanMinus2pct, NoNewsAlphaSmallerThanMinus4pct
)
```

##### Loading 250 most covered by news entities

```python
from blvresearch.core.SP500output import USA_HIGH_NEWS_COVERAGE_250
```

##### Event 2a: many news and alpha > 2%

```python
detect_mnagt2 = EventDetector(ManyNewsAlphaGreaterThan2pct)
```

```python
events_mnagt2 = dict()
for entity_id in USA_HIGH_NEWS_COVERAGE_250:
    entity_data = DATA[entity_id]
    events_mnagt2[entity_id] = detect_mnagt2.run(entity_id, entity_data)
```

```python
list_mnagt2 = EventList.from_dict(events_mnagt2)
```

```python
list_mnagt2
```

```python
pds_mnagt2 = pd.Series(events_mnagt2)
pds_mnagt2 = pds_mnagt2.apply(len)
pds_mnagt2.sort()
```

```python
pds_mnagt2[pds_mnagt2 > 0]
```

```python
plt.hist(pds_mnagt2, bins=range(0,12))
```

```python
df_mnagt2 = create_dataframe_of_daily_effects(list_mnagt2, 'alpha', 20)
plt.figsize(14,8)
daily_means = df_mnagt2.mean(axis=1)
daily_means.cumsum().plot()
df_mnagt2.mean(axis=1)
```

```python
mean_test(list_mnagt2)
```

```python
plot_histogram_of_alphas(list_mnagt2)
```

```python
np.mean([e.concat_data.series_after("alpha", lag=0, length=1) for e in list_mnagt2])
```
