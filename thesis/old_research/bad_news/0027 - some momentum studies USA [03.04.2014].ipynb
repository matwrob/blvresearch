##### Event3: 2-week negative momentum with big drop before

```python
class NegativeMomentumOverLast2WeeksWithBigDrop(NewsEvent):

    def cumulative_alpha_before(self, days):
        return self.alphas_before(days).sum()
    
    def number_of_trading_days_before(self, days):
        ret = self.concat_data.series_before('alpha', lag=1, length=days)
        return len(ret)
    
    def alphas_before(self, days):
        return self.concat_data.series_before('alpha', lag=1, length=days)

    def alpha_drop_before(self, days, drop_size):
        tmp = self.alphas_before(days)
        if any(tmp[tmp < drop_size]):
            return True
        return False

    def triggers(self):
        if (self.cumulative_alpha_before(10) < -0.05 and
            self.number_of_trading_days_before(10) == 10 and
            self.alpha_drop_before(10, -0.05)):
            return True
        return False
```

```python
detector3 = EventDetector(NegativeMomentumOverLast2WeeksWithBigDrop)
```

```python
events3 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector3.run(entity_id, entity_data, META[entity_id])
    if len(entity_events) > 0:
        events3[entity_id] = entity_events
```

```python
events3_wo_consecutive20 = dict()
for entity_id, el in events3.items():
    events3_wo_consecutive20[entity_id] = el.remove_consecutive_events(20)
```

##### Event 4: 2-week positive momentum with big jump before

```python
class PositiveMomentumOverLast2WeeksWithBigJump(NewsEvent):

    def cumulative_alpha_before(self, days):
        return self.alphas_before(days).sum()
    
    def number_of_trading_days_before(self, days):
        tmp = self.alphas_before(days)
        return len(tmp.dropna())
    
    def alphas_before(self, days):
        return self.concat_data.series_before('alpha', lag=1, length=days)

    def alpha_drop_before(self, days, drop_size):
        tmp = self.alphas_before(days)
        if len(tmp[tmp < drop_size]) > 0:
            return True
        return False

    def alpha_jump_before(self, days, jump_size):
        tmp = self.alphas_before(days)
        if len(tmp[tmp > jump_size]) > 0:
            return True
        return False

    def triggers(self):
        if (self.cumulative_alpha_before(10) > 0.05 and
            self.number_of_trading_days_before(10) == 10 and
            self.alpha_jump_before(10, 0.05)):
            return True
        return False
```

```python
detector4 = EventDetector(PositiveMomentumOverLast2WeeksWithBigJump)
```

```python
events4 = dict()
for entity_id, entity_data in DATA.items():
    entity_events = detector4.run(entity_id, entity_data, META[entity_id])
    if len(entity_events) > 0:
        events4[entity_id] = entity_events
```
