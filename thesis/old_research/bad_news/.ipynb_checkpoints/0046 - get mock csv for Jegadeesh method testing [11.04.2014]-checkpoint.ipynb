```python
from concat.core.utils import save_object
```

```python
from blvresearch.core.research_data import (
    get_research_data_by_entity
)
```

```python
entities = ['000YMS-E',  # Wal-Mart Stores, Inc.
            '002Q0G-E',  # Exxon Mobil Corp.
            '000DP6-E',  # Chevron Corp.
            '000C7F-E',  # Apple, Inc.
            '000CS1-E',  # Berkshire Hathaway, Inc.
            '000KYG-E',  # General Electric
            '000Y8N-E',  # Valero Energy
            '000KN2-E',  # Ford Motor Co.
            '069J8N-E',  # AT&T, Inc.
            '000LNN-E']  # Hewlett-Packard Co.
```

```python
start_date = '2011-01-01'
end_date = '2013-12-31'
```

```python
def criteria(sec_meta):
    if sec_meta['avg_vol'] > 1000 and sec_meta['market_cap'] > 1000000:
        return True
    return False
```

```python
DATA = get_research_data_by_entity(
    entities, criteria, start_date, end_date
)
```

```python
save_object(DATA, 'mock_jegadeesh.pickle')
```
